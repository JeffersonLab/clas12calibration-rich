```python
import numpy
import glob, os
from stl.mesh import Mesh
from stl import mesh
from mpl_toolkits import mplot3d
from matplotlib import pyplot
import math

Name_Stl='201_2.stl'
reduction=270
```

```python
your_mesh = mesh.Mesh.from_file('/Users/gangelini/work/JLAB_RICH/GasVolume/RICH_s4.stl')
# Create a new plot
figure = pyplot.figure()
axes = mplot3d.Axes3D(figure)
axes.add_collection3d(mplot3d.art3d.Poly3DCollection(your_mesh.vectors))

# Auto scale to the mesh size
scale = your_mesh.points.flatten()
axes.auto_scale_xyz(scale, scale, scale)

# Show the plot to the screen
pyplot.show()

```

```python
data = numpy.zeros(len(your_mesh), dtype=Mesh.dtype)
num = -4027.4675
tangent = math.tan(60 * math.pi / 180)
deltay= reduction*tangent
print(tangent)
print(deltay)
newvalue = num+reduction
for index, item in enumerate(your_mesh):
    new_vector = numpy.where(item == num, newvalue, item)
    print(new_vector)
    data['vectors'][index]=new_vector.reshape(3,3)
new_mesh = Mesh( data, remove_empty_areas=True)
```
