```python
import numpy
import glob, os
from stl.mesh import Mesh
from stl import mesh
from mpl_toolkits import mplot3d
from matplotlib import pyplot
import math
scale_factor=0.997
```

```python
os.chdir("/Users/gangelini/work/JLAB_RICH/Finalrich/")
for file in glob.glob("*.stl");
    Name_Stl=file;
    your_mesh = mesh.Mesh.from_file(Name_Stl)
    copy_mesh= mesh.Mesh.from_file(Name_Stl)
    volume, cog, inertia = your_mesh.get_mass_properties()
    your_mesh.translate([-cog[0],-cog[1],-cog[2]])
    data = numpy.zeros(len(your_mesh), dtype=Mesh.dtype)
    for index, item in enumerate(your_mesh):
        new_vector = scale_factor*item
        data['vectors'][index]=new_vector.reshape(3,3)
    new_mesh = Mesh( data, remove_empty_areas=False)
    new_mesh.translate([cog[0],cog[1],cog[2]])
    volume2, cog2, inertia2 = new_mesh.get_mass_properties()
    print("Position of the center of gravity old (COG) = {0}".format(cog))
    print("Position of the center of gravity new (COG) = {0}".format(cog2))
    new_mesh.save("/Users/gangelini/work/JLAB_RICH/Finalrich/Small/"+Name_Stl)
```
